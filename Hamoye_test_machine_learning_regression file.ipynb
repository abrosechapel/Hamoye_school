# Test Stage B: Machine Learning: Regression - Predicting Energy Efficiency of Buildings

```python
#Importing packages
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
#Importing dataset
df = pd.read_csv('energydata_complete.csv')
df.head(10)
```

## Question 12

```python
#linear regression model
model = LinearRegression()

#T2 = Celsius and T6 = temperature outside the building 
X, y = df['T6'], df['T2']
X = X[:, None]

#fit regression model
model.fit(X, y)

#calculate R-squared of regression model
r_squared = model.score(X, y)

#view R-squared value
print(r_squared)
```

The coefficient of determination is 0.642. The 64% of the response variable  is explained by the model.

## Question 13

```python
# Prepare the normalization tool
scaler = MinMaxScaler()

#drop columns
df_n=df.drop(columns=[ 'date', 'lights'])

# Normalize the data 
normalised_df = pd.DataFrame(scaler.fit_transform(df_n), columns=df_n.columns)
features_df=normalised_df.drop(columns=[ 'Appliances'])
target = normalised_df[ 'Appliances' ]

#train-test set split

x_train, x_test, y_train, y_test = train_test_split(features_df, target,
test_size= 0.3 , random_state= 42 )

#set the model
linear_model = LinearRegression()

#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#predictions
predicted_y = linear_model.predict(x_test)

#MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_y)
round(mae, 2 ) 
```

## Question 14

```python
#Residual Sum of Squares 
rss = np.sum(np.square(y_test - predicted_y))
round(rss, 2 ) 
```

## Question 15

```python
# Root Mean Squared Error 

rmse = np.sqrt(mean_squared_error(y_test, predicted_y))
round(rmse, 3) 
```

## Question 16

```python
#Coeficient of Determination 

r2_score = r2_score(y_test, predicted_y)
round(r2_score, 2) 
```

## Question 17

```python
linear_weights = pd.Series(linear_model.coef_, x_train.columns).sort_values()
linear_weightsdf = pd.DataFrame(linear_weights).reset_index()
linear_weightsdf.columns = [ 'Features' , 'Linear_Model']
linear_weightsdf['Linear_Model']=linear_weightsdf['Linear_Model'].round(3)
linear_weightsdf
```

The lowest: RH_2, T_out , the highest: RH_1, T3. The highest: TH1 and the lowest: RH_2.
